## Define random walk class

# Example of data:

xyz_positions = [(x_1, y_1, z_1), (x_2, y_2, z_2),..., (x_i, y_i, z_i)]

total_steps = [(step_1, step_1_direction),(step_2, step_2_direction),...,(step_i, step_i_direction)]

```python
class RAND_WALK:
    def __init__(self):
        """
        Initiate the direction map so random decision can be mapped to a particular direction
        Initiate the current position to start at 0,0,0
        """
        self.step_direction_map = {
            0:"x",
            1:'y',
            2:'z'
        }
        self.position = {
            "x":0,
            'y':0,
            'z':0
        }
        
    def random_walk(self, n_step:int):
        """
        Inputs: 
            n_step: Number of steps to take for random walk
        Returns:
            self.position: list of (x,y,z) points for each step of random walk
            self.total_steps: list of steps taken and direction [(step, step_direction)]
        """
        # Initialize total steps and xyz position lists
        self.total_steps = []
        self.xyz_positions = [(0,0,0)]
        
        # Loop over n_step input value
        for idx in range(n_step):
            # Choose a random int from 0 to 2
            random_choice = random.randint(0,2)
            # Map random choice to a direction: Gives random direction
            random_direction = self.step_direction_map[random_choice]
            # Update the position dictonary by a random step size betwen -0.5 and 0.5
            self.position[random_direction] += random.uniform(-1/2, 1/2)
            # Append the updated position and direction to the total steps list
            self.total_steps.append(
                (
                    self.position[random_direction],
                    random_direction
                )
            )
            # Append position dictionary values to xyz positions list
            self.xyz_positions.append([values for values in self.position.values()])

    def random_walk_v2(self, n_step:int):
        """
        Inputs: 
            n_step: Number of steps to take for random walk
        Returns:
            self.position: list of (x,y,z) points for each step of random walk
            self.total_steps: list of steps taken and direction [(step, step_direction)]
        """
        # Initialize xyz position lists
        self.xyz_positions = [(0,0,0)]
        # Loop over n_step input value
        for idx in range(n_step):
            # Update the position dictonary by a random step size betwen -0.5 and 0.5
            self.position["x"] += random.uniform(-1/2, 1/2)
            self.position["y"] += random.uniform(-1/2, 1/2)
            self.position["z"] += random.uniform(-1/2, 1/2)

            # Append position dictionary values to xyz positions list
            self.xyz_positions.append([values for values in self.position.values()])
        
```

```python
W = 10000
N_1 = 1
N_2 = 2
N_3 = 3
N_5 = 5


total_walks_1 = []
total_walks_2 = []
total_walks_3 = []
total_walks_5 = []

for walk in range(W):
    # Initialize RAND_WALK class
    random_walk_1 = RAND_WALK()
    random_walk_2 = RAND_WALK()
    random_walk_3 = RAND_WALK()
    random_walk_5 = RAND_WALK()

    # Call random_walk function to update total steps and xyz position
    random_walk_1.random_walk_v2(N_1)
    random_walk_2.random_walk_v2(N_2)
    random_walk_3.random_walk_v2(N_3)
    random_walk_5.random_walk_v2(N_5)

    # Append the last xyz position from the N-step random walk
    total_walks_1.append(random_walk_1.xyz_positions[-1])
    total_walks_2.append(random_walk_2.xyz_positions[-1])
    total_walks_3.append(random_walk_3.xyz_positions[-1])
    total_walks_5.append(random_walk_5.xyz_positions[-1])
```

```python
# Pulling out X values from xyz results for each step
x1 = [x for x,_,_ in total_walks_1]
x2 = [x for x,_,_ in total_walks_2]
x3 = [x for x,_,_ in total_walks_3]
x5 = [x for x,_,_ in total_walks_5]
```
