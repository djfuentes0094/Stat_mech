```python
import numpy as np
import matplotlib.pyplot as plt

from math import factorial, pi
```

```python
length= 10000

x_list = [np.random.random() for x in range(length)]
y_list = [np.random.random() for x in range(length)]

comb_list = [x + y for x,y in zip(x_list, y_list)]
```

```python
step_size = 0.01
counter_list = {}
for x in np.arange(0.01,2,step_size):
    if x not in counter_list:
        counter_list[round(x,2)]=0
    for point in comb_list:
        if round(x,2) - step_size <= point < round(x,2) + step_size:
            counter_list[round(x,2)]+=1
counter_list = [counter_list[idx] for idx in counter_list]
```

```python
plt.title("Probability Distribution for X + Y Random Numbers")
plt.bar([x for x in np.arange(0.01,2,step_size)],counter_list)
```
