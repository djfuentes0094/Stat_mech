```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Finding divergence of logistic map
### f(x) = 4 u x(1-x)

```python
mu = .9
x_init_1 = .4
epsilon = 1e-8
x_init_2 = x_init_1 + epsilon
x_values_1 = []
x_values_2 = []

for idx in range(200):
    x_final_1 = 4*mu*x_init_1*(1-x_init_1)
    x_final_2 = 4*mu*x_init_2*(1-x_init_2)
    x_values_1.append(x_final_1)
    x_values_2.append(x_final_2)
    x_init_1 = x_final_1
    x_init_2 = x_final_2
    
plt.scatter([x for x in range(len(x_values_1))],x_values_1, marker='x')
plt.scatter([x for x in range(len(x_values_2))],x_values_2, marker='o', color='r')
plt.grid()
```

```python
seperation = [abs(x-y) for x,y in zip(x_values_1, x_values_2)]
seperation_point = [x > .1 for x in seperation]
```

```python
idx = [i for i, x in enumerate(seperation_point) if x][0]
```
