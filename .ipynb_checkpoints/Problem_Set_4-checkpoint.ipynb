```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_init_1 = .75 + 1e-16
x_init_2 = .75 + 1e-8
x_values_1 = []
x_values_2 = []

for idx in range(200):
    x_final_1 = 4*x_init_1*(1-x_init_1)
    x_final_2 = 4*x_init_2*(1-x_init_2)
    x_values_1.append(x_final_1)
    x_values_2.append(x_final_2)
#     print(f"x{idx+1}= {x_final}")
    x_init_1 = x_final_1
    x_init_2 = x_final_2
    
plt.scatter([x for x in range(len(x_values_1))],x_values_1, marker='x')
plt.scatter([x for x in range(len(x_values_2))],x_values_2, marker='o', color='r')
# plt.yscale('log')
plt.grid()
```
